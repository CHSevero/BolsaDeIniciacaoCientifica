```python
import math

class Ponto:
    def __init__(self, x, y):
        self.lat = x
        self.lon = y
        self.id = 0

def distancia(p,q):
    return math.sqrt(math.pow(p.lat-q.lat,2)+math.pow(p.lon-q.lon,2))

def vizinhaca(bd,p,eps):
    v = list()
    for i in bd:
        if(distancia(i,p))<=eps:
            v.append(i)
    return v

def juncao(v,t):
    for p in t:
        if not(p in v):
            v.append(p)
    return v

def dbscan(bd,eps, minpts):
    lista_de_resultado = list()
    pontos_visitados = list()
    vizinhos1 = list()
    
    for p in bd:
        if not(p in pontos_visitados):
            pontos_visitados.append(p)
            vizinhos1 = vizinhaca(bd,p,eps)
            if len(vizinhos1)>=minpts:
                for q in vizinhos1:
                    if not(q in pontos_visitados):
                        pontos_visitados.append(q)
                        vizinhos2 = vizinhaca(bd,q,eps)
                        if len(vizinhos2)>=minpts:
                            vizinhos1 = juncao(vizinhos1,vizinhos2)
                lista_de_resultado.append(vizinhos1)
    return lista_de_resultado

# generating sampl data
from sklearn.datasets.samples_generator import make_blobs
centers = [[5, 5], [0, 0], [1, 5],[5, -1]]
X, labels_true =make_blobs(n_samples=500, n_features=5, centers=centers, cluster_std=0.9, center_box=(1, 10.0), shuffle=True, random_state=0)
D = list()
for i in X:
    D.append(Ponto(i[0],i[1]))
clusters = dbscan(D,0.5,10)
c = 1
for j in clusters:
    print(c)
    c+=1
    for q in j:
        print(q.lat, q.lon, end=" ")
    print()
```
