```python
import math
import psycopg2

# Connect to an existing database
conn = psycopg2.connect("dbname='bolsa' user='postgres' host='localhost' password='severo' port='5432'")

# Open a cursor to perform database operations
cur = conn.cursor()

# Execute a command: this creates a new table
cur.execute("select * from pontos")
bd = cur.fetchall()




# Pass data to fill a query placeholders and let Psycopg perform
# the correct conversion (no more SQL injections!)
#>>> cur.execute("INSERT INTO test (num, data) VALUES (%s, %s)",
#...      (100, "abc'def"))

# Query the database and obtain data as Python objects
#>>> cur.execute("SELECT * FROM test;")
#>>> cur.fetchone()
#(1, 100, "abc'def")

# Make the changes to the database persistent
#>>> conn.commit()

# Close communication with the database
cur.close()

conn.close()

def distancia(a,b):
     return math.sqrt(math.pow(a[0]-b[0],2)+math.pow(a[1]-b[1],2))

def vizinhaca(conj, p, eps):
    vizinhos = set()
    for q in conj:
        if(distancia(p,q) <= eps):
            vizinhos.add(q)
    return vizinhos

conj = [[1,2],[3,2],[5,4],[8,9]]*3
clusters = list()
visitado = set()

eps = 0.001

MinPts = 4
test = list()
n = 0
for p in conj:
    sementes = set()
    cluster = set()
    if not(p in visitado):
        sementes = vizinhaca(conj,p,eps)
        if len(sementes)<MinPts:
            visitado.add(p)
            test.append(-1)
        else:
            test.append(n)
            cluster.add(p)
            visitado.add(p)
            sementes.remove(p)
            sementes2 =set()
            for q in sementes:
                test.append(n)
                cluster.add(q)
                visitado.add(q)
                sementes2 = vizinhaca(conj,p,eps)
                if len(sementes2) >= MinPts:
                    sementes.update(sementes2)
                    sementes.remove(q) 
    
    if len(cluster) > 0:
        n+=1
        clusters.append(cluster) 

nc = 1
for i in clusters:
    print("Cluster %d: "%nc,i)
    print("\n")
    nc+=1

for j in test:
    print(j, end=' ')
```
